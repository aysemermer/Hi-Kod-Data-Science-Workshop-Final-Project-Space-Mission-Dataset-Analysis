```python
unique_companies = data["Company"].unique()
print("Number of unique companies:", len(unique_companies))

unique_rockets = data["Rocket"].unique()
print("Number of unique values in the 'Rocket' column:", len(unique_rockets))
```

```python
data.head(10)
```

```python
total_launches = len(data['MissionStatus'].dropna())
successful_rockets = len(data[data['MissionStatus'] == 1])
print("Total number of launches:", total_launches)
print("Total number of successful rocket launches:", successful_rockets)
```

```python
labels = ['Total Launches', 'Successful Launches']
values = [total_launches, successful_rockets]
plt.bar(labels, values, color=['orange', 'blue'])
plt.title('Total and Successful Rocket Launches')
plt.ylabel('Number of Launches')
plt.grid(True)
plt.show()
```

```python
rocket_counts = data.groupby('Company')['MissionStatus'].agg(['count', 'sum'])
rocket_counts.columns = ['Total Rockets', 'Successful Rockets']
rocket_counts.plot(kind='bar', figsize=(15, 6))

plt.title('Rockets Sent by Each Company')
plt.xlabel('Company')
plt.ylabel('Number of Launches')
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# Line plot grafiği
rocket_counts = data.groupby('Company')['MissionStatus'].agg(['count', 'sum'])
rocket_counts.columns = ['Total Rockets', 'Successful Rockets']
plt.figure(figsize=(15, 6))
plt.plot(rocket_counts.index, rocket_counts['Total Rockets'], marker='o', linestyle='-', color='blue', label='Total Rockets')
plt.plot(rocket_counts.index, rocket_counts['Successful Rockets'], marker='o', linestyle='-', color='green', label='Successful Rockets')
plt.title('Rockets Sent by Each Company')
plt.xlabel('Company')
plt.ylabel('Number of Launches')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
success_by_year = data[data['MissionStatus'] == 1].groupby('Year')['MissionStatus'].count()
plt.figure(figsize=(10, 6))
plt.plot(success_by_year.index, success_by_year.values, marker='o', linestyle='-', color='blue')
plt.title('Successful Rocket Launches Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Successful Launches')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
rocket_company_table = data[['Rocket', 'Company']].drop_duplicates().reset_index(drop=True)
print(rocket_company_table)
```

```python
rocket_company_table = data[['Rocket', 'Company']].drop_duplicates().reset_index(drop=True)
colors = ['orange', 'blue', 'green']
plt.figure(figsize=(15, 8))
rocket_company_table['Company'].value_counts().plot(kind='bar', color=colors)
plt.title('Number of Rockets by Company')
plt.xlabel('Company')
plt.ylabel('Number of Rockets')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
```

```python
first_30_companies = data['Company'].value_counts().index[:30]
last_32_companies = data['Company'].value_counts().index[30:]

data_first_30 = data[data['Company'].isin(first_30_companies)]
data_last_32 = data[data['Company'].isin(last_32_companies)]


unique_rocket_counts_first_30 = data_first_30.groupby('Company')['Rocket'].nunique()
unique_rocket_counts_last_32 = data_last_32.groupby('Company')['Rocket'].nunique()


plt.figure(figsize=(15, 8))
colors = ['orange', 'blue', 'green']

plt.subplot(2, 1, 1)
unique_rocket_counts_first_30.plot(kind='bar', color=colors)
plt.title('Number of Unique Rockets by Company (First 30)')
plt.xlabel('Company')
plt.ylabel('Number of Unique Rockets')
plt.xticks(rotation=90)
plt.grid(True)

plt.subplot(2, 1, 2)
unique_rocket_counts_last_32.plot(kind='bar', color=colors)
plt.title('Number of Unique Rockets by Company (Last 32)')
plt.xlabel('Company')
plt.ylabel('Number of Unique Rockets')
plt.xticks(rotation=90)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
successful_missions = data[data['MissionStatus'] == 1]

first_30_companies = data['Company'].value_counts().index[:30]
last_32_companies = data['Company'].value_counts().index[30:]

successful_missions_first_30 = successful_missions[successful_missions['Company'].isin(first_30_companies)]
successful_missions_last_32 = successful_missions[successful_missions['Company'].isin(last_32_companies)]

mission_counts_first_30 = successful_missions_first_30['Company'].value_counts()
mission_counts_last_32 = successful_missions_last_32['Company'].value_counts()

plt.figure(figsize=(15, 8))
colors = ['orange', 'blue', 'green']
plt.subplot(2, 1, 1)
mission_counts_first_30.plot(kind='bar', color=colors)
plt.title('Number of Successful Missions by Company (First 30)')
plt.xlabel('Company')
plt.ylabel('Number of Successful Missions')
plt.xticks(rotation=90)
plt.grid(True)

plt.subplot(2, 1, 2)
mission_counts_last_32.plot(kind='bar', color=colors)
plt.title('Number of Successful Missions by Company (Last 32)')
plt.xlabel('Company')
plt.ylabel('Number of Successful Missions')
plt.xticks(rotation=90)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
RVSN_USSR_missions = data[data['Company'] == 'RVSN USSR']

successful_RVSN_USSR_missions = RVSN_USSR_missions[RVSN_USSR_missions['MissionStatus'] == 1]
failed_RVSN_USSR_missions = RVSN_USSR_missions[RVSN_USSR_missions['MissionStatus'] == 0]
RVSN_USSR_all_missions = pd.concat([successful_RVSN_USSR_missions, failed_RVSN_USSR_missions])

unique_rockets_RVSN_USSR = data[data['Company'] == 'RVSN USSR']['Rocket'].nunique()
print("Total number of unique rockets launched by RVSN USSR:", unique_rockets_RVSN_USSR)
total_rockets = len(RVSN_USSR_missions)
print("Total number of rockets launched by RVSN USSR:", total_rockets)

display(RVSN_USSR_all_missions)

plt.figure(figsize=(5, 5))
RVSN_USSR_all_missions['MissionStatus'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Number of Successful and Failed Missions by RVSN USRR')
plt.xlabel('Mission Status')
plt.ylabel('Number of Missions')
plt.xticks(ticks=[0, 1], labels=['Successful', 'Failed'], rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()

total_successful_RVSN_USSR_missions = successful_RVSN_USSR_missions.shape[0]
total_failed_RVSN_USSR_missions = failed_RVSN_USSR_missions.shape[0]

print("Total number of successful missions by RVSN USSR:", total_successful_RVSN_USSR_missions)
print("Total number of failed missions by RVSN USSR:", total_failed_RVSN_USSR_missions)
```
